```python
#Libraries
import yfinance as yf

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import datetime
import time

from ta.momentum import RSIIndicator
```

```python
# prompt: download the VIX index for the last 10 years and plot that on the chart

# Download VIX data for the last 10 years
vix = yf.download("^VIX", period="20y")

# Plot the VIX index
plt.figure(figsize=(12, 6))
plt.plot(vix['Close'])
plt.title('VIX Volatility Index (10 Years)')
plt.xlabel('Date')
plt.ylabel('VIX Value')
plt.grid(True)
plt.show()
```

```python
# Calculate the daily change
vix['Daily Change'] = vix['Close'].diff()

# Plot the daily change
plt.figure(figsize=(12, 6))
plt.plot(vix['Daily Change'])
plt.title('Daily Change of VIX Volatility Index (10 Years)')
plt.xlabel('Date')
plt.ylabel('Daily Change')
plt.grid(True)
plt.show()
```
